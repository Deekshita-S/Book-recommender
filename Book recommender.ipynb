```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
books=pd.read_csv('Books.csv')
ratings=pd.read_csv('Ratings.csv')
users=pd.read_csv('Users.csv')
```

```python
booksRatings=books.merge(ratings,on='ISBN')
booksRatings.head()
```

Retain only those users that have rated more than 200 books

```python
#fetching index of only those users that have rated a min of 200 books by boolean indexing
a=booksRatings.groupby(['User-ID']).count()['Book-Rating']>200
User200=a[a].index 
```

```python
filtered_rating=booksRatings[booksRatings['User-ID'].isin(User200)]
```

Retain only those books that have been rated by more than 50 users

```python
b=filtered_rating.groupby(['Book-Title']).count()['Book-Rating']>50
Book50=b[b].index 
finalDF=filtered_rating[filtered_rating['Book-Title'].isin(Book50)]
```

```python
finalDF
```

```python
BookVsUserRating=finalDF.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
BookVsUserRating.fillna(0,inplace=True)
BookVsUserRating
```

```python
similarityScore=cosine_similarity(BookVsUserRating)
```

```python
def recommend(book):
    #fetch index from the given book name
    idx=np.where(BookVsUserRating.index==book)[0][0]
    dist=similarityScore[idx]
    l=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6] # reverse -> to sort in dec order, key = lmbda => to sort based on the second value in the tuple, here the sec val in the tuple is the dist and the first is the index position
    #[1:6] we start at 1 and not at 0 as in the similarity matrix the first entry will the dist between the movie itself thus we want similarity from the second index in the sorted list
    for i in l:
         print(BookVsUserRating.index[i[0]])
```

```python
recommend('The Da Vinci Code')
```
